```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('titanic.csv')
```

# Q1: Import plotly.express library

```python
import warnings
warnings.filterwarnings("ignore")
import plotly.express as px 
```

# Q2: Create a 3D scatter plot represents fare, sibsp and age columns
- Reset the opacity to 0.4
- The color argumnet should be based on pclass column.
- The symbol argumnet should be based on sex column.

```python
df.head(2)
```

```python
fig = px.scatter_3d(df,x="fare",y="sibsp",z="age",color="pclass",symbol="sex",opacity=0.4)
fig.show()
```

# Q3: Create a parallel coordinates plot 
- Let the dimentions be fare, age and survived.
- The color argumnet should be based on age column.
- Reset the color continuous midpoint to be 40.

```python
px.parallel_coordinates(df,dimensions=["fare", "age" , "survived"] ,color = "age",color_continuous_midpoint=40)
```

# Q4: Create a sunburst plot represents alive, sex and pclass column
- The values argument should be based on age column.

```python
fig = px.sunburst(df,path=['alive','sex' ,'pclass'],values="age")
fig
```
